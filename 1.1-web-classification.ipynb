```python
# Import necessary libraries
import requests
from bs4 import BeautifulSoup

# URL of the webpage
url = "https://www.acledabank.com.kh/kh/eng/ps_crhousingloan?t=p"

# Fetch the webpage content
response = requests.get(url)
webpage_content = response.text

# Parse the HTML content
soup = BeautifulSoup(webpage_content, 'html.parser')

soup

# # Extract the div with class 'main'
# main_div = soup.find('div', class_='main')

# main_div
```

```python
# Import necessary libraries
from langchain_community.llms import Ollama

# Initialize the LLaMA 3.1 model
llm = Ollama(model="llama3.1")

# Define the extraction prompt
prompt = f"""
Find the house loan information in the HTML and extract the following:
1. Features of the house loan
2. Requirements for the house loan
3. Website URL

HTML Text:
{webpage_content}
"""

# Run the model to extract information
response = llm.invoke(prompt)

# Display the results
print("Extracted Information:")
print(response)
```
